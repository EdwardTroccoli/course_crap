```python
import cfpack as cfp
from cfpack.defaults import *
import matplotlib.pyplot as plt
import numpy as np
from uncertainties import *
from uncertainties.umath import *
from scipy.signal import find_peaks
```

```python
dat_1a = cfp.read_ascii("/Users/edwardtroccoli/Desktop/ANU - PhB/Courses/3rd Year - Sem 2/PHYS3032/Labs/XRD/DATA/cu_kbr_1a.txt")
dat_1b = cfp.read_ascii("/Users/edwardtroccoli/Desktop/ANU - PhB/Courses/3rd Year - Sem 2/PHYS3032/Labs/XRD/DATA/mo_kbr_1b.txt")
dat_2a = cfp.read_ascii("/Users/edwardtroccoli/Desktop/ANU - PhB/Courses/3rd Year - Sem 2/PHYS3032/Labs/XRD/DATA/cu_LiF_2a.txt")
```

```python
peaks_1a, crap_1a = find_peaks(dat_1a["Impulse"], height=2e2, prominence=1.5e1, distance = 14) 
peaks = [r"$K^{(1)}_{\beta}$",r"$K^{(1)}_{\alpha}$",r"$K^{(2)}_{\beta}$",r"$K^{(2)}_{\alpha}$",r"$K^{(3)}_{\beta}$",r"$K^{(3)}_{\alpha}$"]
color = ["magenta", "magenta", "blue", "blue", "green", "green"]
for peak in range(len(peaks_1a[1:])):
    cfp.plot(x = dat_1a['Crystal angle'][peaks_1a[1:]][peak]-4, 
            y = dat_1a['Impulse'][peaks_1a[1:]][peak],
            text = peaks[peak], color = color[peak]
            )
cfp.plot(x = dat_1a['Crystal angle'][peaks_1a[1:]], y = dat_1a['Impulse'][peaks_1a[1:]], label = "Maxima",
         ylog = True,
         marker='x', linewidth = 2, type="scatter", color = "red")
cfp.plot(x = dat_1a['Crystal angle'], y = dat_1a['Impulse'], color = "goldenrod", label = "Intensity")
cfp.plot(xlabel = r"Scattering Angle $\vartheta$ (deg.)", ylabel = r"Impulse (arb. units)",
         ylog = True, legend_loc = "best", show = True, save = "Cu_KBr_spectra.pdf")
```

```python
dat_1b['Impulse'] = dat_1b['Impulse'] - 173
peaks_1b, crap_1b = find_peaks(dat_1b["Impulse"], height=3e2, prominence=0.75e2, distance = 7) 
peaks_1b = np.concatenate([peaks_1b[:2],peaks_1b[4:]])
peaks = [r"$K^{(1)}_{\beta}$",r"$K^{(1)}_{\alpha}$",r"$K^{(2)}_{\beta}$",r"$K^{(2)}_{\alpha}$",r"$K^{(3)}_{\beta}$",r"$K^{(3)}_{\alpha}$"]
color = ["magenta", "magenta", "blue", "blue", "green", "green"]
dy = [100, 100, 100, 250]
for peak in range(len(peaks_1b)):
    cfp.plot(x = dat_1b['Crystal angle'][peaks_1b][peak] - 1.5, 
            y = dat_1b['Impulse'][peaks_1b][peak]+dy[peak],
            text = peaks[peak], color = color[peak]
            )
cfp.plot(x = dat_1b['Crystal angle'][peaks_1b], y = dat_1b['Impulse'][peaks_1b], label = "Maxima",
         ylog = True,
         marker='x', linewidth = 2, type="scatter", color = "red")
cfp.plot(x = dat_1b['Crystal angle'], y = dat_1b['Impulse'], color = "goldenrod", label = "Intensity")
cfp.plot(xlabel = r"Scattering Angle $\vartheta$ (deg.)", ylabel = r"Impulse (arb. units)",
         ylog = True, legend_loc = "best", show = True, ylim = [1e0, 4e3], save = "Mo_KBr_spectra.pdf")
```

```python
peaks_2a, crap_2a = find_peaks(dat_2a['Impulse'], height=5e1, prominence=4e1, distance = 20) 
peaks = [r"$K^{(1)}_{\beta}$",r"$K^{(1)}_{\alpha}$",r"$K^{(2)}_{\beta}$",r"$K^{(2)}_{\alpha}$",r"$K^{(3)}_{\beta}$",r"$K^{(3)}_{\alpha}$"]
color = ["magenta", "magenta", "blue", "blue", "green", "green"]
for peak in range(len(peaks_2a[1:])):
    cfp.plot(x = dat_2a['Crystal angle'][peaks_2a[1:]][peak]-4, 
            y = dat_2a['Impulse'][peaks_2a[1:]][peak]-10,
            text = peaks[peak], color = color[peak]
            )
cfp.plot(x = dat_2a['Crystal angle'][peaks_2a[1:]], y = dat_2a['Impulse'][peaks_2a[1:]], label = "Maxima",
         ylog = True,
         marker='x', linewidth = 2, type="scatter", color = "red")
cfp.plot(x = dat_2a['Crystal angle'], y = dat_2a['Impulse'], color = "goldenrod", label = "Intensity")
cfp.plot(xlabel = r"Scattering Angle $\vartheta$ (deg.)", ylabel = r"Impulse (arb. units)",
         ylog = True, legend_loc = "best", show = True, ylim = [1e0, 8e3], save = "Cu_LiF_spectra.pdf")
```

```python
deg_to_rad = ufloat(np.pi/180, 0) 
ang1_deg = [ufloat(val, 0.1) for val in dat_1a["Crystal angle"][peaks_1a[1:]]] 
ang2_deg = [ufloat(val, 0.1) for val in dat_1b["Crystal angle"][peaks_1b]]
ang3_deg = [ufloat(val, 0.1) for val in dat_2a["Crystal angle"][peaks_2a[1:]]]
ang1 = [val * deg_to_rad for val in ang1_rad]
ang2 = [val * deg_to_rad for val in ang2_rad]
ang3 = [val * deg_to_rad for val in ang3_rad]
```

```python
# Cu_KBr energies

d, c, h = ufloat(3.29e-10,1e-13), ufloat(3e8,0), ufloat(6.64e-34,0)

n = [1,1,2,2,3,3]

wavelength = [2*d*sin(theta)/order for theta, order in zip(ang1, n)]

Cu_KBr_energies = np.array([h*c/lmbd for lmbd in wavelength])/(1.6e-16)

for i in range(len(n)):
    cfp.print(f"Energy for the peak {peaks[i]} ", Cu_KBr_energies[i])
```

```python
# Mo_KBr energies

d, c, h = ufloat(3.29e-10,1e-13), ufloat(3e8,0), ufloat(6.64e-34,0)

n = [1,1,2,2]

wavelength = [2*d*sin(theta)/order for theta, order in zip(ang2, n)]

Mo_KBr_energies = np.array([h*c/lmbd for lmbd in wavelength])/(1.6e-16)

for i in range(len(n)):
    cfp.print(f"Energy for the peak {peaks[i]} ", Mo_KBr_energies[i])
```
